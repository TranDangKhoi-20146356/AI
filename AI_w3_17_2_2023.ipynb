```python
# load library
import pandas as pd
# create dataframe
dataframe = pd.DataFrame()
# add columns
dataframe['Name'] = ['Lemon','Apple','Banana']
dataframe['Price'] = [5,8,7]
dataframe['License'] = [False,True, True]

# show data
dataframe
```

```python
dataframe.shape
```

```python
df = dataframe
df.info()
```

```python
df.describe() # liet ke tat ca cac cot trong bang
```

```python
# loc iloc
df.loc[(df.Price == 5)]# tim vi tri co gia bang 5 loc(location)
```

```python
df.loc[(df.Price == 5)|(df.Name == 'Banana')]#tim vi tri cai co gia =5 va ten la banana
```

```python
df.loc[(df.Price == 5)| (df.Name == 'Banana'), ['Name','Price']]
```

```python
# create row data
new = pd.Series(['Orange',10,True], index=['Name','Price','License'])
# append row ( thêm vào một cái hàng)
dataframe.append(new, ignore_index=True)# phớt lờ các danh mục vd như name, price, license
```

```python
# append a column
dataframe['Location'] = ['USA','VN','SIN']
dataframe
```

```python
# delete a row
dataframe.drop(1, axis=0).head(10)# xóa index 1, axis 1;0 là cột;hàng, head(10): hiển thị 10 dòng 
```

```python
# delete a column
dataframe.drop('Location', axis=1).head()#(cột cần xóa, axis=1(xóa cột)).hiển thị 5 dữ liệu 
```

```python
# replace
dataframe['Price'].replace(5,7).head(5)
```

```python
# rename columns
dataframe.rename(columns={'Price':'Cost'}).head()
```
