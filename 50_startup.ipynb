```python
import pandas as pd
url = '/content/50_Startups.csv'
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation =  data.corr()#
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot = True)#biểu đồ nhiệt
```

```python
sns.pairplot(data)#
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X= pd.DataFrame(data['Profit']) , y= data['R&D Spend'])  #fit day du lieu
print('he so intercept la:',model.intercept_)
print('he so coeffient la:',model.coef_)




y_pred=model.predict(X=pd.DataFrame(data['Profit']))
data.plot(kind='scatter',x='Administration',y='R&D Spend',figsize=(9,9),color='black')
plt.plot(data['Profit'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Profit'],data['Profit']**2]).T
model.fit(X,y= data['R&D Spend'])
print ('intercept cua mo hinh :', model.intercept_)
print ('coeffient cua mo hinh :', model.coef_)
print('do chinh xac cua mo hinh:',model.score(X,y= data['R&D Spend']))


import numpy as np
range = np.arange(0,200000,1)
X_range= pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter',x ='Profit',y = 'R&D Spend',figsize=(9,9),color = 'black')
plt.plot(range, y_pred,color='blue')
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Profit'],data['Profit']**2]).T
model.fit(X,y=data['R&D Spend'])
print('Intercept cua mo hinh:',model.intercept_)
print('coeffient cua mo hinh:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y=data['R&D Spend']))




```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['Profit'],data['Profit']**2,data['Profit']**3]).T
model.fit(X,y=data['R&D Spend'])
print('Intercept của mô hình:',model.intercept_)
print('Coeffient của mô hình:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X=pd.DataFrame([data['Profit'],data['Profit']**2,data['Profit']**3]).T,y=data['R&D Spend']))
```
