```python
import pandas as pd
url = '/content/Advertising.csv'
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation =  data.corr()#
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot = True)#biểu đồ nhiệt
```

```python
sns.pairplot(data)#biểu đồ trực quan hóa mối quan hệ theo cặp giữa các biến trong dữ liệu tập dữ liệu bằng cách sử dụng các biểu đồ và biểu đồ phân tán.
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X= pd.DataFrame(data['Unnamed: 0']) , y= data['sales'])  #fit day du lieu
print('he so intercept la:',model.intercept_)
print('he so coeffient la:',model.coef_)




y_pred=model.predict(X=pd.DataFrame(data['Unnamed: 0']))
data.plot(kind='scatter',x='Unnamed: 0',y='sales',figsize=(9,9),color='black')
plt.plot(data['Unnamed: 0'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Unnamed: 0'],data['sales']**2]).T
model.fit(X,y= data['sales'])
print ('intercept cua mo hinh :', model.intercept_)
print ('coeffient cua mo hinh :', model.coef_)
print('do chinh xac cua mo hinh:',model.score(X,y= data['sales']))


import numpy as np
range = np.arange(0,200000,1)
X_range= pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter',x ='Unnamed: 0',y = 'sales',figsize=(9,9),color = 'black')
plt.plot(range, y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Unnamed: 0'],data['sales']**2,data['sales']**3]).T
model.fit(X,y= data['sales'])
print ('intercept cua mo hinh :', model.intercept_)
print ('coeffient cua mo hinh :', model.coef_)
print('do chinh xac cua mo hinh:',model.score(X,y= data['sales']))


import numpy as np
range = np.arange(0,200000,1)
X_range= pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter',x ='Unnamed: 0',y = 'sales',figsize=(9,9),color = 'black')
plt.plot(range, y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Unnamed: 0'],data['sales']**2,data['sales']**3,data['sales']**4]).T
model.fit(X,y= data['sales'])
print ('intercept cua mo hinh :', model.intercept_)
print ('coeffient cua mo hinh :', model.coef_)
print('do chinh xac cua mo hinh:',model.score(X,y= data['sales']))


import numpy as np
range = np.arange(0,200000,1)
X_range= pd.DataFrame([range,range**2,range**3,range**4]).T
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter',x ='Unnamed: 0',y = 'sales',figsize=(9,9),color = 'black')
plt.plot(range, y_pred,color='blue')
```
